```python
#Declaring Variables and imports
from manim import *
import numpy as np


vec_a_color = RED_D
vec_b_color = GREEN_D
vec_c_color = BLUE_D

x_color = PINK
y_color = ORANGE
```

## Definindo os Vetores

```python
# Define the vectors
vector_a = [2,2,0]
vector_b = [2,-3,0]

vector_sum = np.array(vector_a) + np.array(vector_b)
```

## Criando a animação

```python
# Create and run the animation

%%manim -qm -v WARNING VectorSum

class VectorSum(Scene):
  def construct(self):
    # Define animation variables
    origin_dot = Dot(ORIGIN)
    number_plane = NumberPlane(
      axis_config={"include_numbers":True}
    )
    
    vector1 = Arrow(ORIGIN, vector_a, buff=0)
    vector2 = Arrow(ORIGIN, vector_b, buff=0)
    vector3 = Arrow(ORIGIN, vector_sum, buff=0)
    
    vector1.stroke_color = vec_a_color
    vector2.stroke_color = vec_b_color
    vector3.stroke_color = vec_c_color

    sum_text = MathTex(
        '\overrightarrow{a}+\overrightarrow{b}=\overrightarrow{c}'
        )

    vec_text_1 = MathTex(
        '\overrightarrow{a}=['+f"{vector_a[0]},{vector_a[1]}]"
        )

    vec_text_2 = MathTex(
        '\overrightarrow{b}=['+f"{vector_b[0]},{vector_b[1]}]"
        )

    vec_text_3 = MathTex(
        '\overrightarrow{c}=[ '+f"{vector_a[0]} + {vector_b[0]}, {vector_a[1]} + {vector_b[1]} ]"
        )
    vec_text_4 = MathTex(
        '[ '+f"{vector_a[0]}+{vector_b[0]}, {vector_a[1]}+{vector_b[1]}]=[{vector_sum[0]},{vector_sum[1]} ]"
        )
    
    origin_text = Text("(0, 0)").next_to(origin_dot, DOWN)

    tip1_text = Text(f"({vector_a[0]}, {vector_a[1]})").next_to(vector1.get_tip())
    tip2_text = Text(f"({vector_b[0]}, {vector_b[1]})").next_to(vector2.get_tip())
    tip3_text = Text(f"({vector_sum[0]}, {vector_sum[1]})").next_to(vector3.get_tip())

    vector_animation = ApplyMethod(vector2.shift, vector1.get_end())

    text_animation = ApplyMethod(sum_text.shift, UP*3+LEFT*5)
    text_animation1 = ApplyMethod(vec_text_1.shift, UP*2+LEFT*5)
    text_animation2 = ApplyMethod(vec_text_2.shift, UP+LEFT*5)

    # Animations

    self.play(Create(sum_text))
    
    self.wait(1)

    self.play(text_animation)

    self.wait(1)

    self.play(Create(vec_text_1))
    
    self.wait(1)

    self.play(text_animation1)
    
    self.wait(1)

    self.play(Create(vec_text_2))
    
    self.wait(1)

    self.play(text_animation2)
    
    self.wait(1)
    
    self.play(Create(vec_text_3))
    self.wait(3)
    self.play(ReplacementTransform(vec_text_3, vec_text_4))
    
    self.wait(3)

    self.remove(sum_text,vec_text_1,vec_text_2,vec_text_3, vec_text_4)

    self.play(GrowFromCenter(origin_dot))
    self.add(origin_text)
    self.play(Create(number_plane), lag_ratio = 0.25)
    
    self.wait(1)
    
    self.remove(origin_text)
    
    self.wait(1)
    
    self.play(Create(vector1))
    self.play(Create(tip1_text))
    
    self.wait(1)
    
    self.play(Create(vector2))
    self.play(Create(tip2_text))
    
    self.wait(2)
    
    self.play(Create(vector3))
    self.play(Create(tip3_text))
    
    self.wait(1)
    
    self.remove(tip1_text,tip2_text, tip3_text)
    
    self.wait(1)
    
    self.play(vector_animation)

    self.wait(5)
    
```
